```
from fastcore.utils import * 

def calc1a(inp):
    lines = inp.splitlines()
    xs, ys = zip(*[line.split() for line in lines])
    xis = L(xs).map(int).sorted()
    yis = L(ys).map(int).sorted()
    res = 0
    for x, y in zip(xis, yis):
        res += abs(x - y)
    return res

calc1a(inp)
```

Yes! NumPy is perfect for this. It can make the code both more concise and faster.

Key benefits here:
- `np.loadtxt` or array operations can parse the input directly
- Vectorized operations: instead of a loop, you can do `np.abs(xis - yis).sum()` in one line
- For large inputs, NumPy's C-based operations are much faster than Python loops

Want to try rewriting `calc1a` using NumPy? A hint: `np.array()` can convert your lists, and arithmetic operations work element-wise on arrays.

```
from numpy import array, sort
import numpy as np
```

```
inp
```

```
xs, ys = array([o.split() for o in inp.splitlines()], int).T
xs, ys
```

```
np.abs(sort(xs) - sort(ys)).sum()
```

```
def calc1a_np(inp):
    xs, ys = array([o.split() for o in inp.splitlines()], int).T
    return np.abs(sort(xs) - sort(ys)).sum()
    
```
